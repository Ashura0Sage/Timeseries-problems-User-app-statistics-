```python
#Loading data for ARIMA model
x_train = df_login[df_login.index <datetime.datetime(2017, 1, 20)]
x_test = df_login[df_login.index >=datetime.datetime(2017, 1, 20)]
```

```python
print(x_train.shape, x_test.shape)
```

```python
#Checking the data if stationery or not..by making the variance constant and which is not dependent on timeseries.
from statsmodels.tsa.stattools import adfuller

def stationarity_test(data):
    dftest = adfuller(data.Login, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print(dfoutput)
    plt.figure(figsize=(16, 7))
    plt.plot(data.index, data.Login)
    plt.show()

stationarity_test(x_train)
```

```python
# Null Hypothesis: data are non-stationary
# large p-values are indicative of non-stationarity, and small p-values suggest stationarity.
# Using the usual 5% threshold, differencing is required if the p-value is greater than 0.05.
```

```python
## Now taking the log (To make variance constant) and check for stationary again 
log_train = x_train
log_train = log_train["Login"].apply(lambda x: math.log(x+1))
log_train = pd.DataFrame(log_train)
stationarity_test(log_train)
```

```python
# The data is still highly non-stationary. But the variance became constant.
#Removing seasonality.
first_diff = log_train["Login"] - log_train["Login"].shift(1)
first_diff = first_diff.fillna(0)
first_diff = pd.DataFrame(first_diff)
stationarity_test(first_diff)
```

```python
#using seasonal differentation for SARIMA model

seasonal_data_differenced = log_train["Login"] - log_train["Login"].shift(12)
seasonal_data_differenced = seasonal_data_differenced.fillna(0)
seasonal_data_differenced = pd.DataFrame(seasonal_data_differenced)
stationarity_test(seasonal_data_differenced)
```

```python
from statsmodels.tsa.stattools import acf, pacf
lag_acf = acf(first_diff, nlags=20)
lag_pacf = pacf(first_diff, nlags=20, method='ols')

plt.figure(figsize=(16, 7))
#Plot ACF: 
plt.subplot(121) 
plt.plot(lag_acf, marker="o")
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(first_diff)),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(first_diff)),linestyle='--',color='gray')
plt.title('Autocorrelation Function')


#Plot PACF:
plt.subplot(122)
plt.plot(lag_pacf, marker="o")
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(first_diff)),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(first_diff)),linestyle='--',color='gray')
plt.title('Partial Autocorrelation Function')
plt.tight_layout()
```

```python
# q – The lag value where the ACF chart crosses the upper confidence interval for the first time. 
# If you notice closely, in this case q=1,q=7 and q=14 has very high positive correlation 

#whereas 

# p – The lag value where the PACF chart crosses the upper confidence interval for the first time. 
# If you notice closely, in this case p=1 has +ve correlation and p=20 and p = 5.5 has high negitive correlation.
```

```python
## AR Model p=1, d=1, q=0 
from statsmodels.tsa.arima_model import ARIMA
model = ARIMA(log_train, order=(1, 1, 0))  
results_AR = model.fit(disp=-1)  
```

```python
plt.figure(figsize=(16, 7))
plt.plot(x_train.index, x_train.values, color="lightblue")
plt.plot(x_test.index, x_test.values, color="green")

## Forecasting 
pred = pd.DataFrame(results_AR.forecast(len(x_test))[0])
pred.columns = ["yhat"]
pred.index = x_test.index

## Converting from log to normal value
pred["yhat"] = pred["yhat"].apply(lambda x: math.exp(x)-1)

## Measuring error. 
measure = math.pow(mean_squared_error(x_test.values, pred.values), 0.5)
print(measure)
plt.plot(pred.index, pred.fillna(0).values, color="red")
plt.show()
```

```python
#using month as a feature..!!
df_login['month'] = [df_login.index[i].month for i in range(len(df_login))]
```

```python
df_login["Jan"] = df_login["month"].apply(lambda x: 1 if x == 1 else 0)

df_login["Dec"] = df_login["month"].apply(lambda x: 1 if x == 12 else 0)
df_login["Feb"] = df_login["month"].apply(lambda x: 1 if x == 2 else 0)
```

```python
df_login["lag1"] = df_login.Login.shift(1) 

## Using the trend
df_login["Trend"] = [i for i in range(len(df_login))]
df_login = df_login.fillna(0)
df_login.head()

#df_login = df_login.drop(['month'],1)
```

```python
#Split into train and test.
x_train = df_login[df_login.index <datetime.datetime(2017, 1, 20)]
x_test = df_login[df_login.index >=datetime.datetime(2017, 1, 20)]
```

```python
#using log of y instead of direct 'y' due to multiplicative effect 
x_train["Login"] = x_train["Login"].apply(lambda x: math.log(x+1))
x_train["lag1"] = x_train["lag1"].apply(lambda x: math.log(x+1))

x_test["lag1"] = x_test["lag1"].apply(lambda x: math.log(x+1))
x_train.head()
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression().fit(x_train[["Jan", "Dec", "Feb", "lag1",  "Trend"]], x_train.Login)
pred = model.predict(x_test[["Jan", "Dec", "Feb", "lag1",  "Trend"]])
measure = math.pow(mean_squared_error(x_test.Login, [math.exp(i)-1 for i in pred]), 0.5)
print("RMSE:", measure)
```

```python
print("model.coef_: {}".format(model.coef_)) 
print("model.intercept_: {}".format(model.intercept_))
```

```python
pred = pd.DataFrame(pred)
pred.columns = ["yhat"]
pred.index = x_test.index
pred.head()

plt.figure(figsize=(16, 7))
plt.plot(x_train.index, [math.exp(i)-1 for i in x_train.Login], color="lightblue")
plt.plot(x_test.index, x_test.Login, color="green")
plt.plot(pred.index, [math.exp(i)-1 for i in pred.values], color="red")
plt.show()
```

```python
#plotting residual
plt.figure(figsize=(16, 7))
plt.scatter(x_test["Login"].values, [math.exp(i)-1 for i in pred["yhat"].values])
plt.xlabel("Actual values")
plt.ylabel("Predicted values")
plt.title("Actual vs Predicted")
```

```python
plt.figure(figsize=(16, 7))
plt.hist([j -  math.exp(i)-1 for i, j in zip(pred["yhat"].values, x_test.Login)])
plt.title("Residual plots")
```

```python
error = [j -  math.exp(i)-1 for i, j in zip(pred["yhat"].values, x_test.Login)]
datetime = x_test.index

plt.figure(figsize=(16, 7))
plt.scatter(datetime, error)
plt.xlabel("datetime")
plt.ylabel("Error")
plt.title("Residual Plots")
```
